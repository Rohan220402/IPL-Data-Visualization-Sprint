```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the IPL dataset
ipl_df = pd.read_csv("ipl_dataset_expanded_900.csv")  # Ensure the CSV is in the same folder
```

```python
# Step 3: Process the data to get total runs by each batsman
player_runs = ipl_df.groupby("batsman")["runs"].sum().sort_values(ascending=False)
```

```python
# Step 4: Extract top 10 run scorers and reset index for plotting
top_10 = player_runs.head(10).reset_index()
top_10.columns = ['Player', 'Total Runs']  # Renaming columns for clarity
```

```python
# Step 5: Set a clean visual style using Seaborn
sns.set_style("whitegrid")
```

```python
# Step 7 (Final): Plot bar chart with proper formatting and no warnings

plt.figure(figsize=(12, 6))  # Set figure size

# Use 'hue' to apply palette without warning, and hide redundant legend
barplot = sns.barplot(data=top_10, x="Player", y="Total Runs", hue="Player", palette="viridis", legend=False)

# Add title and axis labels
plt.title("Top 10 Run Scorers in IPL", fontsize=16, fontweight="bold")
plt.xlabel("Player Name", fontsize=12)
plt.ylabel("Total Runs Scored", fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Show the final plot
plt.show()
```
