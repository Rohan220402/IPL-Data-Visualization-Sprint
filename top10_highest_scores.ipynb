```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the CSV
ipl_df = pd.read_csv("ipl_dataset_expanded_900.csv")
```

```python
# Step 3: Group by 'batsman' and sum up the 'runs'
batsman_runs = ipl_df.groupby("batsman")["runs"].sum().sort_values(ascending=False)
```

```python
# Step 4: Get Top 10 Batsmen
top_10 = batsman_runs.head(10).reset_index()
```

```python
# Step 5: Plot using Seaborn
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.barplot(x="runs", y="batsman", data=top_10, palette="magma")

plt.title("Top 10 Highest Run Scorers in IPL Dataset", fontsize=16)
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Batsman", fontsize=12)
plt.tight_layout()
plt.show()
```
